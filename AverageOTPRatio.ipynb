```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "MBTA_Commuter_Rail.csv"
df = pd.read_csv(file_path)

route_otp = df.groupby("gtfs_route_id")["otp_ratio"].mean().reset_index()
route_otp = route_otp.sort_values(by="otp_ratio", ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.barplot(
    data=route_otp,
    y="gtfs_route_id",
    x="otp_ratio",
    hue="gtfs_route_id",      
    palette="Blues_d",
    color = "blue",
    dodge=False,              
    legend=False               
)
plt.ylabel("Route ID")
plt.xlabel("Average On-Time Performance Ratio")
plt.title("Average OTP Ratio by MBTA Commuter Rail Route (2024)")
plt.xlim(0.8, 1.01)
plt.tight_layout()

plt.savefig('plot.png')

plt.show()
```
